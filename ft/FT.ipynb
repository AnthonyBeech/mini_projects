```python
import numpy as np
import math as mt
import cmath as cmt
import matplotlib.pyplot as plt
```

```python
L=8
N=2**L  #number of samples
f=10         #frequency of input

  
sig = np.zeros((2*N, N), dtype=np.complex) #create empty array


for i in range(N):                                    #define input function
    sig[0,i] = mt.cos(2*mt.pi*f*i/N)
    sig[1,i] = sig[0,i]
    
    
for k in range(N):                        #perform discrete hilbert transform to produce imaginary component
    if (k % 2) == 0:
        for n in range(N):
            if (n % 2) != 0:
                sig[1,k]=sig[1,k]+(2/mt.pi)*sig[0,n]/(k-n)*1j
    else:
        for n in range(N):
            if (n % 2) == 0:
                sig[1,k]=sig[1,k]+(2/mt.pi)*sig[0,n]/(k-n)*1j 
              
                
print(N)
```

```python
for j in range(N):
    for n in range(N):
        sig[2,j]=sig[2,j]+sig[1,n]*cmt.exp(-1j*2*mt.pi*j*n/N)
```

```python
plt.plot(sig[1].real, label="R")
plt.plot(sig[1].imag, label="I")
plt.legend()
plt.show()
```
