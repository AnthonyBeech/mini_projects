```python
import re
import sys
import time
from google_trans_new import google_translator  
n=0
N=0
  
translator = google_translator()  

infile = "casa.xml"
outfile = infile.replace(".xml", "_tran.xml")

delete_list = [
    "</span>", "</p>", """<span style="style_3">""", """<span style="style_1">""","""<span style="style_2">""","""<span style="style_4">""",
     """<span style="span">""", """<span style="italic">""","â™ª"
              ]
with open(infile, encoding="utf8") as fin:
    for line in fin:
        N=N+1


with open(infile, encoding="utf8") as fin, open(outfile, "w+") as fout:
    for line in fin:
        line = line.replace("<br/>", "-")
        for word in delete_list:
            line = line.replace(word, "")
            line = re.sub(r'\<.*?\>','',line, flags=re.DOTALL)
        if not re.match(r'^\s*$', line):
            fout.write(line)
            line1=translator.translate(line, lang_src='es', lang_tgt='en')
            fout.write(line1+ '\n\n')
        n=n+1
        time.sleep(1)
        if n % 100==0:
            time.sleep(60)
        if n % 10==0:
            print(("%d / %d \r") % (n,N))
            sys.stdout.flush()
```
